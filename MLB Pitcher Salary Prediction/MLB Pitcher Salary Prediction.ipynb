```python
import pandas as pd
pd.set_option('display.max_columns',None)
```

# Getting a Pitching stats DF

```python
df = pd.read_csv('Pitching.csv')
df = df[df['yearID'] > 1984]
```

```python
df.shape
```

```python
df=df.sort_values('W',ascending=False)
df.head()
```

```python
df['stint'].value_counts()
```

```python
players = df.groupby('playerID')['stint'].nunique().reset_index()
players
```

```python
df_grouped = df.groupby(['playerID', 'yearID'])['W', 'L', 'G', 'GS',
       'CG', 'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'SO', 'BAOpp',
       'ERA', 'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP'].sum()
df_grouped = df_grouped.reset_index()
df_grouped = df_grouped.sort_values('W',ascending=False).reset_index(drop='index')
```

```python
print(df_grouped.shape)
df_grouped.head(25)
```

# Getting Postseason DF

```python
post_pitch = pd.read_csv('PitchingPost.csv')
post_pitch = post_pitch[post_pitch['yearID']>1984]
post_pitch = post_pitch.groupby(['playerID','yearID'])['W', 'L', 'G', 'GS',
       'CG', 'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'SO', 'BAOpp',
       'ERA', 'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP'].sum()
post_pitch = post_pitch.reset_index()
post_pitch = post_pitch.sort_values('W',ascending=False).reset_index(drop='index')
```

```python
#post_pitch = post_pitch.drop(columns=['round','teamID','lgID'])
print(post_pitch.shape)
post_pitch.head()
```

```python
post_pitch.columns
```

```python
post_pitch = post_pitch.rename(columns = {'playerID':'playerID', 
'yearID': 'yearID', 
'W': 'Post_W', 
'L': 'Post_L', 
'G': 'Post_G', 
'GS': 'Post_GS',
'CG': 'Post_CG', 
'SHO': 'Post_SHO', 
'SV': 'Post_SV', 
'IPouts': 'Post_IPouts',
'H': 'Post_H', 
'ER': 'Post_ER', 
'HR': 'Post_HR', 
'BB': 'Post_BB', 
'SO': 'Post_SO', 
'BAOpp': 'Post_BAOpp', 
'ERA': 'Post_ERA', 
'IBB': 'Post_IBB', 
'WP': 'Post_WP', 
'HBP': 'Post_HBP', 
'BK': 'Post_BK',
'BFP': 'Post_BFP', 
'GF': 'Post_GF', 
'R': 'Post_R', 
'SH': 'Post_SH', 
'SF': 'Post_SF', 
'GIDP': 'Post_GIDP'})
```

```python
merged2 = pd.merge(df_grouped, post_pitch, how='left', on=['playerID', 'yearID'])
print(merged2.shape)
merged2.head()
```

```python
merged2.iloc[0]
```

# Getting Awards DF

```python
awards = pd.read_csv('AwardsPlayers.csv')
awards = awards[awards['yearID'] >1984]
print(awards.shape)
awards.head()
```

```python
cy_young = awards[awards['awardID'] == 'Cy Young Award']
cy_young = cy_young.groupby(['playerID', 'yearID'])['awardID'].count().reset_index()

gold_glove = awards[awards['awardID'] == 'Gold Glove']
gold_glove = gold_glove[gold_glove['notes'] == 'P']
gold_glove = gold_glove.groupby(['playerID', 'yearID'])['awardID'].count().reset_index()
```

```python
cy_young.head()
```

```python
gold_glove.head()
```

```python
merged3 = pd.merge(merged2, cy_young, how='left', on=['playerID', 'yearID'])
merged4 = pd.merge(merged3, gold_glove, how='left', on=['playerID', 'yearID'])
merged4 = merged4.rename(columns = {'awardID_x': 'Cy_Young', 'awardID_y': 'Gold_Glove'})
print(merged4.shape)
merged4.head()
```

# Getting Allstar DF

```python
all_star = pd.read_csv('AllstarFull.csv')
all_star = all_star[all_star['yearID'] > 1984]
all_star = all_star.groupby(['playerID', 'yearID'])['gameID'].count().reset_index()
all_star
```

```python
merged5 = pd.merge(merged4, all_star, how='left', on=['playerID', 'yearID'])
merged5 = merged5.rename(columns = {'gameID': 'All_star'})
print(merged5.shape)
merged5.head()
```

# Getting Names from Master DF

```python
players = pd.read_csv('Master.csv')
players['Name'] = players['nameFirst'] + ' ' + players['nameLast']
players = players[players['finalGame'] > '1984-1-1']
players = players[['playerID', 'throws', 'Name']]
```

```python
players.head()
```

```python
merged6 = pd.merge(merged5, players, how='left', on='playerID')
print(merged6.shape)
merged6.head()
```

# Getting Salary DF

```python
salary = pd.read_csv('Salaries.csv')
salary = salary[salary['yearID']>1984]
salary.shape
```

```python
salary['year'] = salary['yearID'] - 1
salary = salary.drop(columns = 'yearID')
salary = salary.rename(columns = {'year': 'yearID'})
```

```python
salary.head()
```

```python
merged = pd.merge(merged6, salary, on=['playerID', 'yearID'])
data = merged.drop(columns=['teamID','lgID'])
```

```python
print(data.shape)
data.head()
```

# Final DataFrame

```python
data = data.fillna(0)
print(data.shape)
data.head()
```

```python
data.dtypes
```
