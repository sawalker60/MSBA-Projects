### Carefully examine the data for issues. 

````python
#The first thing that needs to be done is convert each column name to lowercase and not have any whitespace
try:
    new_columns = []
    for col in data.columns:
        new_columns.append(col.replace(' ', '_').lower())
    new_columns

    data.columns = new_columns

except Exception as e:
    print(f"An exception was raised: {e}")
````

### Convert columns to their proper data type. Provide the code used to make the change and provide your reasoning in the comments. State which columns changed, how they changed, and why you made the change.

````python
# Convert columns to their proper data type. Provide the code used to make the change and provide your reasoning in the comments. State which columns changed, how they changed, and why you made the change.
try:
    # The 'date' column type was changed to a datetime 
    data['date'] = pd.to_datetime(data['date'], format = '%Y-%m-%d')

    # The zip_code type was changed to a string as we only need to count these instead of using other mathematical computations (sum,multiply,etc.)
    data['zip_code'] = data.zip_code.astype(str)

    # The store_number type was changed to a string as we only need to count these instead of using other mathematical computations (sum,multiply,etc.)
    data['store_number'] = data.store_number.astype(str)

    # The county_number type was changed to a string as we only need to count these instead of using other mathematical computations (sum,multiply,etc.)
    data['county_number'] = data.county_number.astype(str)

    # The vendor_number type was changed to a string as we only need to count these instead of using other mathematical computations (sum,multiply,etc.)
    data['vendor_number'] = data.vendor_number.astype(str)

    # The item_number type was changed to a string as we only need to count these instead of using other mathematical computations (sum,multiply,etc.)
    data['item_number'] = data.item_number.astype(str)

    # The category type was changed to a string as we only need to count these instead of using other mathematical computations (sum,multiply,etc.)
    data['category'] = data.category.astype(str)
    
except Exception as e:
    print(f"An exception was raised: {e}")
````

### Provide code for any data changes that you made to correct the data or to make the data more consistent for analysis. Provide your reasoning for the changes.

````python
# Provide code for any data changes that you made to correct the data or to make the data more consistent for analysis. Provide your reasoning for the changes.
try:
    # For each row with a 'nan' value for 'address', it is filled with the string 'Missing'
    data.address = data.address.fillna('Missing')
    
    # For each row with a 'nan' value for 'city', it is filled with the string 'Missing'
    data.city = data.city.fillna('Missing')
    
    # For each row with a 'nan' value for 'store_location', it is filled with the string 'Missing'
    data.store_location = data.store_location.fillna('Missing')
    
    # For each row with a 'nan' value for 'county', it is filled with the string 'Missing'
    data.county = data.county.fillna('Missing')
    
    # For each row with a 'nan' value for 'category_name', it is filled with the string 'Missing'
    data.category_name = data.category_name.fillna('Missing')
    
    # For each row with a 'nan' value for 'vendor_name', it is filled with the string 'Missing'
    data.vendor_name = data.vendor_name.fillna('Missing')

    # To combine the counties with the same spelling but dividided into lower and upper case values
    data.county = data.county.str.title()
    
except Exception as e:
    print(f"An exception was raised: {e}")
````

### 1. How much liquor was sold by county (bottles sold, volume in gallons, and dollar amount)?

````python
# How much liquor was sold by county (bottles sold, volume in gallons, and dollar amount)?
try:
    bottles_sold_by_county = data.groupby(['county'])['bottles_sold'].sum().sort_values(ascending = False)
    volume_sold_by_county = data.groupby(['county'])['volume_sold_(gallons)'].sum().sort_values(ascending = False)
    dollar_sales_by_county = data.groupby(['county'])['sale_(dollars)'].sum().sort_values(ascending = False)

    print(bottles_sold_by_county.head())
    print()
    print(volume_sold_by_county.head())
    print()
    print(dollar_sales_by_county.head())
    
except Exception as e:
    print(f"An exception was raised: {e}")
````

### 2. Provide a summary of total sales using these categories: Whiskey, Vodka, and Rum. Which category was highest in sales?

````python
# Provide a summary of total sales using these categories: Whiskey, Vodka, and Rum. Which category was highest in sales?

try:
    #Using the contains string function, we can filter the data by category and sum the total sales
    whiskies_data = data[data['category_name'].str.contains('Whiskies')]
    whiskies_sales = whiskies_data['sale_(dollars)'].sum()


    vodka_data = data[data['category_name'].str.contains('Vodka')]
    vodka_sales = vodka_data['sale_(dollars)'].sum()


    rum_data = data[data['category_name'].str.contains('Rum')]
    rum_sales = rum_data['sale_(dollars)'].sum()

    print(f'Whiskey: ${whiskies_sales:.2f} total sales')
    print(f'Vodka:    ${vodka_sales:.2f} total sales')
    print(f'Rum:      ${rum_sales:.2f} total sales')
    print()
    print(f'Whiskey has the highest amount of sales with ${whiskies_sales:.2f} total sales')

except Exception as e:
    print(f"An exception was raised: {e}")
````
